```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
user = pd.read_csv("Data/user.csv")
product = pd.read_csv("Data/product.csv")
quality = pd.read_csv("Data/quality.csv")
order = pd.read_csv("Data/order.csv")
```

```python
# generate us_date table from user
# 
us_date = user.date.value_counts().to_frame()
us_date = us_date.reset_index()
us_date.columns = ['date', 'user_count']
us_date = us_date.sort_values(by=['date']).reset_index(drop=True)
```

```python
us_date.iloc[140:170]
```

```python
us_date.plot(x = 'date', y = 'user_count', figsize=(12,4), label = 'user_count')

# two drops are Luner New Year
# rises before new year
# - look into the terms
```

```python
w1 = us_date['user_count'].iloc[:7].reset_index()
w2 = us_date['user_count'].iloc[7:14].reset_index()
w3 = us_date['user_count'].iloc[14:21].reset_index()
```

```python
plt.figure(figsize=(12,4)) 
plt.plot(w1['user_count'], label = 'week1')
plt.plot(w2['user_count'], label = 'week2')
plt.plot(w3['user_count'], label = 'week3')
plt.legend()
plt.show()
```

```python
m1 = us_date['user_count'].iloc[:30].reset_index()
m2 = us_date['user_count'].iloc[30:60].reset_index()
m3 = us_date['user_count'].iloc[59:90].reset_index()
```

```python
plt.figure(figsize=(12,4)) 
plt.plot(m1['user_count'], label = 'month1')
plt.plot(m2['user_count'], label = 'month2')
plt.plot(m3['user_count'], label = 'month3')
plt.legend()
plt.show()
```

```python
X0 = us_date.user_count
diff0 = list()
for i in range(7, len(X0)):
    value0 = X0[i] - X0[i - 7]
    diff0.append(value0)
plt.figure(figsize=(12,4)) 
plt.plot(diff0)
plt.show()

#every a while there is a jump
```

```python
X1 = us_date.user_count
diff1 = list()
for i in range(30, len(X1)):
    value1 = X1[i] - X1[i - 30]
    diff1.append(value1)
plt.figure(figsize=(12,4)) 
plt.plot(diff1, color = 'orange')
plt.show()
# monthly cycle is removed 
```

```python
X2 = us_date.user_count
diff2 = list()
for i in range(91, len(X2)):
    value2 = X2[i] - X2[i - 91]
    diff2.append(value2)
plt.figure(figsize=(12,4)) 
plt.plot(diff2, color = 'olive')
plt.show()
```

```python
plt.figure(figsize=(12,4)) 
plt.plot(diff0[83:])
plt.plot(diff1[60:])
plt.plot(diff2)
plt.show()
```

```python
# 1.35 years in total, so a yearly cycle is not accurate according to the sample size

X3 = us_date.user_count
diff3 = list()
for i in range(365, len(X3)):
    value3 = X3[i] - X3[i - 365]
    diff3.append(value3)
plt.plot(diff3)
plt.show()
```

```python
from scipy import stats

new_col_num = ['term', 'limit', 'result', 'count', 'term_per', 'limit_per', 'result_per']

nr_rows = 3
nr_cols = 3

fig, axs = plt.subplots(nr_rows, nr_cols, figsize=(nr_cols*3.5,nr_rows*3))

for r in range(0,nr_rows):
    for c in range(0,nr_cols):  
        i = r*nr_cols+c
        if i < len(new_col_num):
            sns.regplot(order_sum2[new_col_num[i]], order_sum2['date'], ax = axs[r][c])
            stp = stats.pearsonr(order_sum2[new_col_num[i]], order_sum2['date'])
            str_title = "r = " + "{0:.2f}".format(stp[0]) + "      " "p = " + "{0:.2f}".format(stp[1])
            axs[r][c].set_title(str_title,fontsize=11)
            
plt.tight_layout()    
plt.show()   
```
